### Monthly sales

```python
# Monthly Sales
df_monthly = df_1[pd.date_range('2019-08-01',periods=12,freq='M')].sum()
df_monthly
```

### Quartely sales

```python
Q3_2019_sum = df_1[pd.date_range('2019-08-01',periods=2,freq='M')].sum().sum()
Q4_2019_sum = df_1[pd.date_range('2019-10-01',periods=3,freq='M')].sum().sum()
Q1_2020_sum = df_1[pd.date_range('2020-01-01',periods=3,freq='M')].sum().sum()
Q2_2020_sum = df_1[pd.date_range('2020-04-01',periods=3,freq='M')].sum().sum()
Q3_2020_sum = df_1[pd.date_range('2020-07-01',periods=1,freq='M')].sum().sum()
print('Q3_2019_sum: ', Q3_2019_sum)
print('Q4_2019_sum: ', Q4_2019_sum)
print('Q1_2020_sum: ', Q1_2020_sum)
print('Q2_2020_sum: ', Q2_2020_sum)
print('Q3_2020_sum: ', Q3_2020_sum)
```

```python
# Top 3 monthly sales
df_monthly_dict = dict(df_monthly)
top_3 = sorted(df_monthly_dict.values(),reverse=True)[:3]
for key, value in df_monthly_dict.items():
    if value in top_3:
        print(key)
```

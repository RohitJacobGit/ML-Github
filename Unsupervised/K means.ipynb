```python
n_samples = 1000
n_centers = 10
X, y = make_blobs(n_samples=n_samples, centers=n_centers)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X[:,0],X[:,1],color='purple',alpha='0.1',cmap=y)
```

```python
fig, ax = plt.subplots(figsize=(6,4))
# plt.figure(figsize=(6,4))
# ax.axis([-15, 15, -15, 15])
ax.scatter(X[:,0],X[:,1],color='purple',alpha='0.1',cmap=y)
init_data = -10 + 20*np.random.rand(n_centers, 2)
center_scatter = ax.scatter(init_data[:,0],init_data[:,1],marker='*',s=100,color='purple',cmap=y)
```

```python
def init():
    return center_scatter,

def animate(i):
    if i == 0:
        center_scatter.set_offsets(init_data)
        center_scatter.set_color('black')
    else:
        kmns = KMeans(n_clusters=n_centers, 
                      max_iter=i)
        kmns.fit(X)
        center_scatter.set_offsets(kmns.cluster_centers_)
        center_scatter.set_color('blue')
    return center_scatter,
    
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=5, 
                               interval=1000, blit=True)

HTML(anim.to_html5_video())
```

```python
kmns = KMeans(n_clusters=n_centers)
kmns.fit(X)
plt.scatter(X[:, 0], X[:, 1], c=kmns.labels_, cmap=plt.cm.cool, alpha=0.1)
plt.scatter(kmns.cluster_centers_[:, 0], kmns.cluster_centers_[:, 1], 
            marker='*', s=100, color='purple')
```

```python

img = rescale(imread('WhatsApp Image 2020-06-17 at 15.21.44.jpeg'), 0.5)

plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.axis('off')
```

```python
cvecs = img.reshape((img.shape[0]*img.shape[1], img.shape[2]))
pd.DataFrame(cvecs).head(10)
```

```python
cvecs.shape
```

```python
import matplotlib.patches as patches
def visualize_centroids(centroids):
    fig, (ax1, ax) = plt.subplots(2, 1)
    ax1.imshow(img)
    ax1.axis('off')
    ax.axis('off')
    ax.axis([0, 0.2*len(centroids)+0.2, 0.0, 0.8])
    for i, cen in enumerate(centroids):
        p = patches.Rectangle((0.1+(i*0.2), 0.1), 0.2, 0.6,
                             facecolor=cen)
        ax.add_patch(p)

def perform_kmeans_and_visualize(n_clusters):
    kmns = KMeans(n_clusters=n_clusters)
    kmns.fit(cvecs)
    plt.figure(figsize=(10, 10))
    visualize_centroids(kmns.cluster_centers_)
```

```python
perform_kmeans_and_visualize(2)
```

```python
perform_kmeans_and_visualize(3)
```

```python
perform_kmeans_and_visualize(4)
```

```python
def reconstruct_image(n_colors=2):
    kmns = KMeans(n_clusters=n_colors)
    kmns.fit(cvecs)
    plt.imshow(kmns.cluster_centers_[kmns.labels_].reshape(img.shape))
    plt.axis('off')

plt.figure(figsize=(10, 10))
reconstruct_image(2)
```

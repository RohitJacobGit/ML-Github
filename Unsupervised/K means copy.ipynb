```python
kmeans_7 = KMeans(7)
model_7 = kmeans_7.fit(data)
label_7 = model_7.predict(data)

fig = plt.figure();
ax = Axes3D(fig)
ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=label_7, cmap='tab10');
```

```python
wcss = []
#scores = []

for center in range(1,12):
    kmeans = KMeans(center, init='k-means++')
    model = kmeans.fit(data)
    wcss.append(model.inertia_)
    #scores.append(abs(model.score(data)))
    
centers = list(range(1,12))
plt.plot(centers,wcss, linestyle='--', marker='o')
```
